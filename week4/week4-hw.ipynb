```python

def FASTAReader(file):
    line = file.readline()
    assert line.startswith('>'), "Not a FASTA file"
    seq_id = line[1:].rstrip('\r\n')
    sequence = []
    line = file.readline()
    sequences = []
    while line:
        if line.startswith('>'):
            sequences.append((seq_id, ''.join(sequence)))
            seq_id = line[1:].rstrip('\r\n')
            sequence = []
        else:
            sequence.append(line.strip())
        
        line = file.readline()
    sequences.append((seq_id, ''.join(sequence)))

    return sequences
```
